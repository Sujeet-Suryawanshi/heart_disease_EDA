# Changing the numerical Data values to cartegorical Data

```python
def cp(row):
    if row == 0:
        return ' Angina'
    elif row == 1:
        return 'Typical_Angina'
    elif row == 2:
        return 'Non_Anginal_Pain'
    elif row == 3 :
        return 'Asymptomatic'
```

```python
# df['cp'].apply(cp)
df['cp'] = df['cp'].apply(cp)
```

```python
Person_having_Heart_Disease = df.age.where(df.target == 1)

sns.histplot(Person_having_Heart_Disease)
plt.title('No. of people having Heart disease')
plt.show()
```

```python
Healthy_Persons = df.age.where(df.target == 0)
sns.histplot(Healthy_Persons)
plt.title ('People with no Heart Disease')
plt.show()
```

```python
def heart_disease(row):
    if row == 0:
        return 'Absence'
    elif row==1:
        return 'Presence'
```

```python
df.groupby('target').size()
```

```python
df['heart_disease'] = df['target'].apply(heart_disease)
df.head()
```

# Correlations

The Corrolation heatmap Analyzed all the factors

```python
corr = df.drop(['cp','heart_disease','Sex'],axis=1).corr()
plt.subplots(figsize=(15, 10))
cmap = sns.diverging_palette(250, 30, as_cmap=True)
sns.heatmap(corr, cmap=cmap, center=0, square=True, linewidths=1)
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.drop(['cp','heart_disease','Sex'],axis=1).corr(), annot=True, linewidth=1)
plt.show()
```
